## Create New Column from Existing Column

```python
nba = pd.read_csv("nba.csv")
nba.head()
```

```python
nba["Age"]
```

```python
nba["Age"] + 10
```

```python
nba["Age in a Decade"] = nba["Age"] + 10
nba["Age in a Decade"] = nba["Age"].add(10)
```

```python
nba.head()
```

```python
nba["Salary"]
```

```python
nba["Salary"].sub(5000000)
```

```python
nba["Tax on Salary"] = nba["Salary"] * 0.25
```

```python
nba.head()
```

```python
nba["Weight"]
```

```python
weight_in_kg = nba["Weight"] * 0.4535
```
